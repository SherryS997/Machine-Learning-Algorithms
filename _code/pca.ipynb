```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()
```

```python
mu = np.array([1, 2])
cov = np.array([[1, -0.8], [-0.8, 1]])
X = rng.multivariate_normal(mu, cov, size = 1_000).T
X.shape
```

```python
plt.scatter(X[0, :], X[1, :])
```

```python
X = X - X.mean()
cov = X@X.T/X.shape[1]
eigenvals, eigenvecs = np.linalg.eigh(cov)
```

```python
PC1 = eigenvecs[1].reshape((-1, 1))
PC1
```

```python
proj = (X.T@PC1)*PC1.T
```

```python
x = np.linspace(-4, 3)
plt.figure(figsize=(5,5))
plt.scatter(X[0, :], X[1, :])
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(proj[:, 0], proj[:, 1], color="red", label="Projection")
plt.legend();
```
